## **Def I: Tail Function**</br>
We define the <u>tail function</u> $\bar{F}$ of a distribution $F$ on $\mathbb{R}$ to be given by $\bar{F}(x)=F(x,\infty)$ for all $x$. We describe as a <u>tail property</u> of $F$ any properties which depends only on {$F(x):x \geq x_0$} for any finite $x_0$. We further say that $F$ has right-unbounded support if $\bar{F}(x) > 0$ for all $x$.



## **Def II: Heavy-Tailed**</br>
The distribution $F$ on $\mathbb{R}$ is said to be right heavy tailed if $$\int^{\infty}_{-\infty}e^{\lambda x}F(dx)=\infty\text{ for all }\lambda \text{ > 0}$$
that is , if and only if $F$ fails to possess any positive exponential moment. Otherwise, $F$ is said to be light-tailed. In other words, if the moment-generating function of the probability distribution converges, then the distributuion is not heavy tailed.


## <u>Exercise:</u>
Identify if the Cauchy Distribution is heavy tailed using the above definition. The pdf for the Cauchy distribution is $f(x)=\frac{1}{\pi(1+x^{-2})}$</br>
Let us evaluate the following integral:
$$\int^{\infty}_{-\infty}\text{ }e^{\lambda x}\frac{1}{\pi(1+x^{-2})}$$

To show that the above integral diverges, let us observe $e^{\lambda x}$ and $\frac{1}{\pi(1+x^{-2})}$ as $x$ goes to infinity.




```python
lam = 0.5
x = np.linspace(1,10,100) # x-values
f = lambda x: np.exp(lam * x) # first function
g = lambda x: 1/(np.pi * (1 + x**2)) # second function

exp = f(x) # exponential function values
poly = g(x) # polynomial function values

plt.plot(x,exp,label="Exponential Function",color="blue") # plot exponenial function
plt.plot(x,poly,label="Polynomial Function",color="orange") # plot polynomial function
# Labels fo Graph
plt.xlabel("x")
plt.ylabel("Function Value")
plt.title("Comparison of Exponential and Polynomial Function")
plt.legend()
```

We can see that the exponential function converges to infinity at a faster rate than the polynomial converges to zero. Therefore, this integral will converge to infinity as $x \rightarrow \infty$.

## **Def III: Long-Tailed**</br>
A distribution $F$ on $\mathbb{R}$ is said to be <u>long-tailed</u> if $F$ has right-unbounded support and for any fixed $y>0$,
$$\frac{\bar{F}(x+y)}{\bar{F}(x)} \rightarrow 1\text{ as }x \rightarrow \infty$$
Note: long-tailed distributions are heavy-tailed, however, the above condition implies a degree of smoothness in the ail function $F$ which does not occur in every heavy-tailed distribution.



# Subexponential Distributions
We require a slightly stronger regularity condition than the requirement that such a distribtuion be long-tailed, sonce some heavy-tailed distributions are not. Let $F$ be any distribtution on $\mathbb{R^{+}}$ and let $\xi_1,...,\xi_n$ be independent random variables with common distribution $F$. Then
$$P\{\xi_1+...+\xi_n>x\}$$
$$\geq P\{max(\xi_1,...,\xi_n)>x\}$$
$$=1-F^{n}(x) \sim n\bar{F}(x)\text{ as }x \rightarrow \infty$$
Where $\sim$ means "the ratio of the quantities on either side of theis symbol converges to 1"

Taking $n=2$:
$$\lim_{x\to\infty} inf\frac{\bar{F*F}(x)}{\bar{F}(x)} \geq 2$$
Further, the distribution $F$ on $\mathbb{R^{+}}$ is said to be subexponential if
$$\lim_{x\to\infty}\frac{\bar{F*F}(x)}{\bar{F}(x)}=2$$
If this condition is satisfied, $F$ is heavy-tailed and long-tailed. Subexpoenetiality is also a tail property of the distributon $F$. Generally
$$\lim_{x\to\infty}\frac{F^{*n}(x)}{F(x)}=n$$
Notice $F^{*n}(x)$ is a convolution of $F$ to itself.

This argument is equivalent to the requirement that $P\{max(\xi_1,...,\xi_n)>x\} \sim P\{\xi_1+...+\xi_n>x\}$ as $x \rightarrow \infty$

Interpretation: The only significant way i nwhich the sum $\xi_1+...+\xi_n$ can exceed some large value $x$ is that the maximum of one of the individual random variables $\xi_1,...,\xi_n$ also exceeds $x$. This is known as "The principle of a single big jump".
